```python
# import cell_agents/agents/CASSIA/CASSIA_run/cassia_results_processed.h5ad and view

import anndata as ad
adata = ad.read_h5ad("cassia_results_processed.h5ad")

# show unique value in the first column ofm cassia_results_summary.csv
import pandas as pd
cassia_results_summary = pd.read_csv("cassia_results_summary.csv")
# check if the unique values in the cluster column of adata are the same as the unique values in the first column of cassia_results_summary.csv
clusters_adata = set(adata.obs["cluster"].unique())
clusters_cassia = set(cassia_results_summary[cassia_results_summary.columns[0]].unique())

clusters_adata_int = {int(c) for c in adata.obs["cluster"].unique()}

print("Clusters in adata not in cassia results:", clusters_adata_int - clusters_cassia)
print("Clusters in cassia results not in adata:", clusters_cassia - clusters_adata_int)
```
